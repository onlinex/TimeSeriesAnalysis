```python
def calculate_wape(y_true, y_pred, split_id):
    """
    Возвращает ошибку WAPE модели на основе полученных данных факта и прогноза по дням
    :param y_true: np.array факт
    :param y_pred: np.array прогноз
    :param split_id: int идентификатор сплита
    :return: wape
    """

    BIG_SPLITS = [1, 2, 8]

    if len(y_true) != len(y_pred):
        raise ValueError("Размерности y_true и y_pred не совпадают")

    corr_fact = np.where(y_true == 0, 1, y_true)
    corr_forecast = np.where((y_true == 0) & (y_pred == 0), 1, y_pred)

    if split_id in BIG_SPLITS:
        wape = abs(corr_fact - corr_forecast).sum() / corr_fact.sum()

        if wape > 1:
            return wape

        return wape
    else:
        if y_pred.sum() == 0:
            if y_true.sum() == 0:
                return 0
            else:
                return 1

        acc_day = np.where((corr_fact + corr_forecast) == 0, 1, 0)
        acc_day = np.where((corr_fact >= corr_forecast) & (corr_fact != 0), corr_forecast / corr_fact, acc_day)
        acc_day = np.where((corr_fact < corr_forecast) & (corr_forecast != 0), corr_fact / corr_forecast, acc_day)

        wape = 1 - (acc_day * corr_fact).sum() / corr_fact.sum()

        return wape
```

<h1>PMF</h1>

```python
def getPMF(df, weekdays=[0,1,2,3,4,5,6], normalize=True, cutoff=0.7, split=None):
    # !!! this is костыль, don't touch it please !!! #
    
    if split == 56 and weekdays==[0]:
        cutoff=0.7
    
    # cutoff - % of last available data to be used
    df = df.sort_values(by='datetime', ascending=True)
    # count available values
    vc = df['datetime'].dt.date.value_counts()
    # get only complete ones (96)
    dfc = df[df['datetime'].dt.date.isin(vc[vc==96].index)]
    # pick weekdays
    dfw = dfc[dfc['datetime'].dt.weekday.isin(weekdays)]
    # cutoff if necessary
    dfw = dfw[dfw['datetime'] > dfw['datetime'].quantile(1-cutoff).date()]
    # normalize for each day (convert to pmf)
    if normalize:
        dfw['calls'] = dfw.groupby(dfw['datetime'].dt.date)['calls'].apply(lambda x: x/x.sum())
        # if for some reason sum is 0
        dfw['calls'] = dfw['calls'].fillna(0)
    # average by time
    pmf = dfw.groupby(dfw['datetime'].dt.time)['calls'].mean()
    std = dfw.groupby(dfw['datetime'].dt.time)['calls'].std()
    
    return pmf, std, dfw
```

<h1>Restore Calls per day</h1>

```python
def restore(x, pmfs: dict):
    # if more than % of data is missing
    if x['calls'].isna().sum() > 0.8 * len(x['calls']):
        return None
    # get weekday
    weekday = x['datetime'].dt.weekday.mode().values[0]
    # choose pmf for the weekday
    pmf = pmfs[weekday]
    # get the total daily estimate of from the data that is present
    calls_estimates_total = x['calls'].values / pmf.loc[x['datetime'].dt.time]
    # deal with inf values
    calls_estimates_total.replace([np.inf, -np.inf], np.nan, inplace=True)
    # calculate eastimates for each 15 min
    calls_estimates = calls_estimates_total.mean() * pmf
    # return estimates for nan + true values
    return calls_estimates[x['calls'].isna().values].sum() + x['calls'].sum()
```

```python
def getCalls(df, pmfs=None):
    # get pmf as a dictionary for each weekday
    if not pmfs:
        pmfs = {i:getPMF(df, weekdays=[i])[0] for i in range(7)}
    
    st = df['datetime'].iloc[ 0].replace(hour=0, minute=0, second=0)
    en = df['datetime'].iloc[-1].replace(hour=23, minute=45, second=0)
    # construct index with 15 min intervals
    idx = pd.date_range(st, en, freq='15min', name='datetime')
    # reindex
    dfr = df.set_index('datetime').reindex(idx, fill_value=None).reset_index()
    # apply restoration
    dfd = dfr.groupby(dfr['datetime'].dt.date).apply(restore, pmfs=pmfs).reset_index(name='calls')
    # convert to datetime
    dfd['datetime'] = pd.to_datetime(dfd['datetime'], yearfirst=True, dayfirst=False)
    # sort by datetime
    dfd = dfd.sort_values(by='datetime', ascending=True)
    
    dfd.dropna(inplace=True) # drop nan
    
    return dfd
```

<h1>Add holidays</h1>

```python
def parse_holidays():
    # for each year
    holidays = []
    for year in [2016, 2017, 2018, 2019, 2020, 2021]:
        root = xml.etree.ElementTree.parse('./Holidays/{}.xml'.format(year)).getroot()
        
        for d in root.findall('./days/day'):
            month, day = d.get('d').split('.')
            
            str_date =  str(year) + '-' + month + '-' + day
            
            holidays.append(str_date)
            
    return pd.to_datetime(pd.Series(holidays))

def add_holidays(df):
    # get holidays
    holidays_dates = parse_holidays().dt.date
    # fill values
    df['holiday'] = df['datetime'].dt.date.isin(holidays_dates).astype(int)
    
    return df
```

<h1>Add news</h1>

```python
def add_news(df):
    news = pd.read_csv('./Trends/SberbankNews.csv', header=1, names=['datetime', 'v'])
    news['v'] += pd.read_csv('./Trends/SberbankNews1.csv', header=1, names=['datetime', 'v'])['v']
    news['v'] += pd.read_csv('./Trends/SberbankNews2.csv', header=1, names=['datetime', 'v'])['v']
    # convert to datetime
    news['datetime'] = pd.to_datetime(news['datetime'], yearfirst=True, dayfirst=False)
    # sort by datetime
    news = news.sort_values(by='datetime', ascending=True)
    # reindex with foward fill
    st = news['datetime'].iloc[ 0]
    en = news['datetime'].iloc[-1]
    # construct index
    idx = pd.date_range(st, en, freq='15min', name='datetime')
    # reindex
    news = news.set_index('datetime').reindex(idx, method='ffill').reset_index()
    
    df = pd.merge(df, news, on='datetime')
    return df
```

```python
def add_stock(df):
    ticker = yf.Ticker('SBER.ME')
    
    hist = ticker.history(period='5y', interval='1d')['Close']
    # sort by datetime
    hist = hist.sort_index()
    # reindex with foward fill
    st = hist.index[0]
    en = hist.index[-1]
    # construct index
    idx = pd.date_range(st, en, freq='15min', name='datetime')
    # reindex
    hist = hist.reindex(idx, method='ffill')
    
    df = df.set_index('datetime').join(hist, how='left').reset_index()
    
    return df
```

```python
def genFeatures(df):
    
    X = deepcopy(df)
    try:
        Y = X.pop('calls')
    except:
        Y = None
    
    X['m'] = X['datetime'].dt.month
    X['d'] = X['datetime'].dt.day
    X['dy'] = X['datetime'].dt.dayofyear
    X['y'] = X['datetime'].dt.year
    X['w'] = X['datetime'].dt.weekday
    
    X = X.drop('datetime', axis=1)
    
    return X, Y
```

```python
avg_score = []

tscv = TimeSeriesSplit(n_splits=6, test_size=458)
for train_index, test_index in tscv.split(X):
    print("TRAIN:", len(train_index), "TEST:", len(test_index))
    
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = Y.iloc[train_index], Y.iloc[test_index]
    
    reg = RandomForestRegressor(n_estimators=38, max_depth=7, criterion='mse', random_state=0).fit(X_train, y_train)
    
    print('train mae:', mean_absolute_error(y_train, np.round(reg.predict(X_train))))
    print('valid mae:', mean_absolute_error(y_test,  np.round(reg.predict(X_test))))
    
    avg_score.append(mean_absolute_error(y_test,  np.round(reg.predict(X_test))))
    
print(np.mean(avg_score))
```
